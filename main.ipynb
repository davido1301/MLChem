```python
# Bibliotheken
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import Lasso
from sklearn.model_selection import learning_curve
from sklearn.model_selection import GridSearchCV
import tensorflow as tf
from tensorflow.keras import layers, models
from scipy.stats import zscore 
```

```python
# INPUT & CLEANING

# Werte in geeignetes Input Format überführen 
def clean_value(value):
    try:
        return float(value)
    except ValueError:
        return float(value.replace('*^-', 'e-').replace('*^', 'e'))


# Funktion für das Lesen einer einzelnen Datei des Datasets
def read_qm9_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    num_atoms = int(lines[0].strip())
    
    # Moleküleigenschaften: A, B, C, mu, alpha, homo, lumo, gap, r2, zpve, U0, U, H, G, Cv
    props = lines[1].strip().split()
    
    # GDB raus, Index = Nummer des Files 
    identifier = props[0]
    index = int(props[1])  # 1 las Int gecastet, gdb ignoriert 
    A, B, C, mu, alpha, homo, lumo, gap, r2, zpve, U0, U, H, G, Cv = map(float, props[2:])
    
    # Atomdaten, xyz, Art 
    atoms_data = [line.strip().split() for line in lines[2:2 + num_atoms]]
    columns = ['Element', 'x', 'y', 'z', 'Additional_Property']
    atoms_df = pd.DataFrame(atoms_data, columns=columns)
    
    for col in ['x', 'y', 'z', 'Additional_Property']:
        atoms_df[col] = atoms_df[col].apply(clean_value)
    
    # Zusatzinfo 
    additional_properties = lines[2 + num_atoms].strip().split()
    additional_properties = list(map(float, additional_properties))
    
    return {
        'Identifier': identifier,
        'Index': index,
        'Num_Atoms': num_atoms, # Funktionierts? 
        'A': A,
        'B': B,
        'C': C,
        'mu': mu,
        'alpha': alpha,
        'homo': homo,
        'lumo': lumo,
        'gap': gap,
        'r2': r2,
        'zpve': zpve,
        'U0': U0,
        'U': U,
        'H': H,
        'G': G,
        'Cv': Cv,
        'Atoms_Data': atoms_df,
        'Additional_Properties': additional_properties
    }

# Alle Daten einlesen im Directory Qm9/ 
def read_all_qm9_files(directory):
    all_data = []
    
    for filename in os.listdir(directory):
        if filename.endswith('.xyz'):
            file_path = os.path.join(directory, filename)
            data = read_qm9_file(file_path)
            data['Filename'] = filename
            all_data.append(data)
    
    return all_data

directory = 'QM9/'

# # Wenn noch keine Pickle Datei 
# all_qm9_data = read_all_qm9_files(directory)
# with open('qm9_data.pickle', 'wb') as f:
#    pickle.dump(all_qm9_data, f)

# Mit Pickle Datei: 
with open('qm9_data.pickle', 'rb') as f:
     all_qm9_data = pickle.load(f)




#TEST 
example_data = all_qm9_data[0]
# Allgemeiner Überblick über die Daten
def print_example_data(data):
    print(f"Filename: {data['Filename']}")
    print(f"Identifier: {data['Identifier']}")
    print(f"Index: {data['Index']}")
    print(f"AtomNummer: {data['Num_Atoms']}") 
    print(f"A: {data['A']} GHz")
    print(f"B: {data['B']} GHz")
    print(f"C: {data['C']} GHz")
    print(f"mu: {data['mu']} Debye")
    print(f"alpha: {data['alpha']} Bohr^3")
    print(f"homo: {data['homo']} Hartree")
    print(f"lumo: {data['lumo']} Hartree")
    print(f"gap: {data['gap']} Hartree")
    print(f"r2: {data['r2']} Bohr^2")
    print(f"zpve: {data['zpve']} Hartree")
    print(f"U0: {data['U0']} Hartree")
    print(f"U: {data['U']} Hartree")
    print(f"H: {data['H']} Hartree")
    print(f"G: {data['G']} Hartree")
    print(f"Cv: {data['Cv']} cal/(mol K)")
    
    print('\nAtoms Data:')
    print(data['Atoms_Data'].to_string(index=False))
    
    print('\nAdditional Properties:')
    print(data['Additional_Properties'])

print_example_data(example_data)



# Statistik 
def compute_aggregates(data):
    sums = {'homo': 0, 'lumo': 0, 'gap': 0, 'U': 0, 'H': 0, 'G': 0, 'Cv': 0, 'Num_Atoms': 0}
    sum_squares = {'homo': 0, 'lumo': 0, 'gap': 0, 'U': 0, 'H': 0, 'G': 0, 'Cv': 0, 'Num_Atoms': 0}
    n = len(data)  
    
    for molecule in data:
        for key in sums.keys():
            value = molecule[key]
            sums[key] += value
            sum_squares[key] += value ** 2
            
    averages = {key: sums[key] / n for key in sums.keys()}
    std_devs = {key: np.sqrt(sum_squares[key] / n - (sums[key] / n) ** 2) for key in sum_squares.keys()}
    return averages, std_devs

averages, std_devs = compute_aggregates(all_qm9_data)

print("Averages:")
for key, value in averages.items():
    print(f"{key}: {value:.6f}")

print("\nStandard Deviations:")
for key, value in std_devs.items():
    print(f"{key}: {value:.6f}")

```

## SVM

Die Support-Vektor-Machine (SVM) ist ein supervised-learning Model, welches für Klassifizierung verwendet wird. Hierbei wird jedes Objekt als Vektor eines n-dimensionalen Raums darsgestellt. Am Beispiel eines Datensatzes mit zwei Klassen: das Verfahren sucht nach der besten Linie (mit mehreren Klassen höhere Dimensionen, also Hyperebenen), die die beiden Klassen voneinander unterscheidbar macht und damit trennt. Im linearen Fall wäre dies für zwei Klassen eine gerade Linie. Für nichtlineare Fälle wie im folgenden RBF (Radial Basis Funktion) werden die Datem in einen höherdimensionalen Raum transformiert, in welchen sie linear voneinander trennbar sind. (Kernel-Trick). Ziel ist es dabei, die Klassen optimal zu trennen, sprich eine Entscheidungsgrenze zu finden, die die Klassen eindeutig unterscheidet. SVC = Support Vector Classification.

```python
#! Dauert länger

y_atoms = np.array([data['Num_Atoms'] for data in all_qm9_data])


X_train_atoms, X_test_atoms, y_train_atoms, y_test_atoms = train_test_split(X, y_atoms, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled_atoms = scaler.fit_transform(X_train_atoms)
X_test_scaled_atoms = scaler.transform(X_test_atoms)


pca_atoms = PCA(n_components=2)  # 2D
X_train_pca_atoms = pca_atoms.fit_transform(X_train_scaled_atoms) 

X_test_pca_atoms = pca_atoms.transform(X_test_scaled_atoms)

# SVM
svm_clf = SVC(random_state=42)
svm_clf.fit(X_train_pca_atoms, y_train_atoms)

# num_atoms
y_pred_atoms = svm_clf.predict(X_test_pca_atoms)

# Metrik
accuracy = accuracy_score(y_test_atoms, y_pred_atoms)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test_atoms, y_pred_atoms))


```

# Theorie Lineare Regression

Lineare Regression beschreibt die Beziehung zwischen Labels und Feautures und versucht eine Gerade zu ermitteln bzw. Hyperplane für mehrere Feature, welche die Beziehung am besten darstellt. 

In der Funktion LinearRegression() von scikit-learn wird die Ordinary Least Squares Linear Regression verwendet.
Dabei wird folgende Form verwendet

$ Y = \beta_{0}  + \beta_{1} X_{1} + ... + \epsilon $

Dabei sind X die Features und Y die Targets bzw. Labels. Das $\epsilon$ stellt dabei einen Fehler Term dar. Die $\beta$ Werte sind die Gewichtungen bzw. Koeffizienten.

Es wird versucht die residuale Summe der Quadrate (RSS) zu minimieren.

$ RSS = \sum_{i}^{n} (Y_{i} - Y_{i, pred})^2 $

woraus sich folgende Form für die optimalen Koeffizienten  $\beta$ in Matrix-Notation berechnen lassen: 
$ \hat{\beta} = (X^{T} X)^{-1} X^{T} Y $






```python
# LIN REGRESSION

features = []
labels = [] 

for data in all_qm9_data:
    features.append([
        data['Num_Atoms'], # data['A'], data['B'], data['C'],
        data['mu'], data['alpha'], data['r2'], # data['zpve'],
        data['H'], data['G'], data['Cv']
        , data['lumo'], data['gap'], data['U0'], data['U']
    ]) 
    labels.append([
        data['homo']
    ])

X = np.array(features)
y = np.array(labels)

# FÜR HOMO --------------------------------------------------------------------------------------
y_homo = y[:, 0]  # HOMO

X_train, X_test, y_train, y_test = train_test_split(X, y_homo, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LinearRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE) HOMO: {mse:.2f}")
print(f"Mean Absolute Error (MAE) HOMO: {mae:.2f}")
print(f"R²-Score HOMO: {r2:.2f}")

# tatsächlich vs. vorhergesagten Werte
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # 45° Linie
plt.xlabel('Tatsächliche Werte')
plt.ylabel('Vorhergesagte Werte')
plt.title('Tatsächliche vs. Vorhergesagte Werte für HOMO')
plt.show()

# Berechnung der Baseline
baseline_mean = np.mean(y_train)  # Mittelwert des Outputs Trainingsset
baseline_std = np.std(y_train)    # Standardabweichung des Outputs Trainingsset

# Baseline-MSE
baseline_mse = mean_squared_error(y_test, np.full_like(y_test, baseline_mean))
baseline_mae = mean_absolute_error(y_test, np.full_like(y_test, baseline_mean))

print(f"Baseline MSE (Mittelwert-Vorhersage): {baseline_mse:.2f}")
print(f"Baseline MAE (Mittelwert-Vorhersage): {baseline_mae:.2f}")
print(f"Standardabweichung des Outputs: {baseline_std:.2f}")

# # Lernkurve erstellen
# train_sizes, train_scores, test_scores = learning_curve(model, X, y_homo, cv=5, scoring='neg_mean_squared_error',
#                                                         train_sizes=np.linspace(0.1, 1.0, 10), random_state=42)

# train_scores_mean = -train_scores.mean(axis=1)
# test_scores_mean = -test_scores.mean(axis=1)

# plt.figure(figsize=(8, 6))
# plt.plot(train_sizes, train_scores_mean, label='Trainingsfehler')
# plt.plot(train_sizes, test_scores_mean, label='Testfehler')
# plt.axhline(y=baseline_mse, color='r', linestyle='--', label='Baseline MSE (Mittelwert)')
# plt.xlabel('Anzahl der Trainingsdaten')
# plt.ylabel('Fehler (MSE)')
# plt.title('Lernkurve für HOMO (Lineare Regression)')
# plt.legend()
# plt.show()

```

# Erweitere Regressionsmethoden Theorie
Hier für: Lasso (Least Absolute Shrinkage and Selection Operator) 

Die Grundidee der linearen Regression bleibt die Gleiche. Auch hier wird wieder eine lineare Gleichung der Form 
$ Y = \beta_{0}  + \beta_{1} X_{1} + ... + \epsilon $
verwendet.

Das Konzept der linearen Regression (s.o.) wird um eine Penalty erweitert. Es wird eine Loss Function minimiert der folgenden Form

$ Loss Function = SSE + \lambda \sum_{j=1}^p | \beta_{j} | $
Lambda ist ein Hyperparameter, der die Regulierung kontrolliert. Beta sind weiterhin die Gewichtungen und p sind die Anzahl der Regressionsparameter. Durch die Minimierung der Loss-Function werden die Beta Werte ermittelt. 

Der Vorteil von Lasso gegenüber der linearen Regression ergibt sich, wenn zu viele unabhängige Feature verwendet werden, was dazu führt, dass die lineare Regression overfittet wird, d.h., dass das Modell gut für die Testdaten verwendet werden kann, jedoch zu spezifisch ist für andere Datensätze und dort versagt.

# NN Neuronales Netzwerk 

Neuronale Netzwerke bestehen aus Schichten und Knoten und ahmen die Funktionsweise des Zusammenwirkens von Neuronen nach. 
Die erste Schicht ist die Input-Schicht, sprich die Input Daten. Jeder Knoten bzw. im Sinne der Analogie Neuron besteht aus aus einem Merkmal bzw. der Feauture des Modells. 
Die letzte Schichte ist die Output Schicht, also die Vorhersage des Netzwerks. Zwischen des Input-Layers und Output-Layers befinden sich die sogenannten Hidden-Layers. Jedes Neuron in einer Schicht ist mit jedem weiteren Neuron der vorgegangen und nachfolgenden Schicht verbunden. Jede Verbindung hat eine bestimmte Gewichtung, welche bestimmt, wie stark das Signal weitergegeben wird. 

Jedes Neuron in der Hidden-Layer berechnet eine gewichtete Summe der Eingabewerte (bzw. weiter gereichten Werte des vorgegangenen Layers), welche als Input für eine Aktivierungsfunktion dient (Hier: ReLU). Diese bestimmt ob das Neuron aktiviert wird und Informationen weiter gibt oder nicht. 

Während des Lernprozesses werden mehrere Schritte durchlaufen. 


Forward Pass: 
Die Eingabedaten werden von Input-Layer über Hidden-Layer zum Output-Layer geleitet und eine Vorhersage wird erzeugt. 

Berechnung des Fehlers: 
Die Vorhersage wird mit dem tatsächlichen Wert verglichen und ein Loss-Algorithmus angewendet, um den Fehler zu bestimmen. Hierbei eignen sich Metriken wie MAE. 

Anschließend folgt die Rückpropagation: 
Hier erfolgt das Signal durch den Output-Layer über Hidden-Layer zum Input-Layer. Dabei werden die Gewichtungen der Verbindungen der Neuronen angepasst und optimiert. 

Durch diesen iterativen Prozess lernt das NN seine Gewichte anzupassen, sodass immer bessere Vorhersagen ermittelt werden können. 






```python
# NN HOMO ------------------------------------------------------------------------------

X = np.array(features)  
y_homo = np.array([data['homo'] for data in all_qm9_data])  # HOMO


X_train, X_test, y_train, y_test = train_test_split(X, y_homo, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1))  

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

history = model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)

test_loss, test_mae = model.evaluate(X_test_scaled, y_test)
print(f"Mean Absolute Error (MAE) auf dem Testset: {test_mae:.2f}")

y_pred = model.predict(X_test_scaled)

# Tatsächliche vs. vorhergesagte Werte
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # 45° Linie
plt.xlabel('Tatsächliche Werte')
plt.ylabel('Vorhergesagte Werte')
plt.title('Tatsächliche vs. Vorhergesagte HOMO-Werte (Neuronales Netzwerk)')
plt.show()

# Lernkurve
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Trainingsverlust')
plt.plot(history.history['val_loss'], label='Validierungsverlust')
plt.xlabel('Epoche')
plt.ylabel('MSE Verlust')
plt.title('Lernkurve')
plt.legend()
plt.show()

# Berechnung der Baseline
baseline_mean = np.mean(y_train)  # Mittelwert des Outputs Trainingsset
baseline_std = np.std(y_train)    # Standardabweichung des Output Trainingsset

# Baseline-MSE 
baseline_mse = mean_squared_error(y_test, np.full_like(y_test, baseline_mean))
baseline_mae = mean_absolute_error(y_test, np.full_like(y_test, baseline_mean))

print(f"Baseline MSE (Mittelwert-Vorhersage): {baseline_mse:.2f}")
print(f"Baseline MAE (Mittelwert-Vorhersage): {baseline_mae:.2f}")
print(f"Standardabweichung des Outputs: {baseline_std:.2f}")

#Metriken 
r2 = r2_score(y_test, y_pred)
print(f"R-Squared (R2) on Test Set: {r2:.2f}")

```
